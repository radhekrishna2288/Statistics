```python
# Q1. What is a random variable in probability theory?

'''
A random variable is a variable whose possible values are numerical outcomes of a random phenomenon. There are two
types of random variables: discrete and continuous.

Example: The number of heads in 10 coin flips is a discrete random variable.
Explanation: In this example, the random variable can take on any integer value from 0 to 10, depending on the 
number of heads observed in 10 coin flips.
'''
```

```python
# Q2. What are the types of random variables?

'''
There are two types of random variables: discrete and continuous.

Discrete random variables take on a countable number of distinct values. Continuous random variables take on an
infinite number of possible values.

Example: The number of students in a class is a discrete random variable, while the height of students is a 
continuous random variable.
'''
```

```python
# Q3. What is the difference between discrete and continuous distributions?

'''
Discrete distributions describe the probability of outcomes of a discrete random variable, while continuous 
distributions describe the probability of outcomes of a continuous random variable.

Example: The binomial distribution is a discrete distribution, while the normal distribution is a continuous
distribution.
'''
```

```python
# Q4. What are probability distribution functions (PDF)?

'''
A probability distribution function (PDF) describes the likelihood of different outcomes for a random variable.
For discrete random variables, it is called a probability mass function (PMF), and for continuous random variables,
it is called a probability density function (PDF).

Example: The PDF of a normal distribution is a bell-shaped curve.

Explanation: The PDF provides the probability density of the random variable at each point, and the area under the
curve represents the total probability, which is 1.
'''
```

```python
# Q5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

'''
A cumulative distribution function (CDF) gives the probability that a random variable is less than or equal to a 
certain value, while a probability distribution function (PDF) gives the probability of the random variable taking
on a specific value.

Example: The CDF of a normal distribution is an S-shaped curve.

Explanation: The CDF is the integral of the PDF, and it shows the cumulative probability up to a certain value.
'''
```

```python
# Q6. What is a discrete uniform distribution?

'''
A discrete uniform distribution is a distribution where all outcomes are equally likely. Each value has the same
probability of occurring.

Example: Rolling a fair six-sided die results in a discrete uniform distribution.

Explanation: Each face of the die (1 through 6) has an equal probability of 1/6.
'''
```

```python
# Q7. What are the key properties of a Bernoulli distribution?

'''
A Bernoulli distribution is a discrete distribution with only two possible outcomes: success (1) and failure (0).
The probability of success is denoted by p, and the probability of failure is 1-p.

Example: Flipping a coin where heads is considered success and tails is considered failure.

Explanation: The probability of getting heads (success) is p, and the probability of getting tails (failure) is
1-p.
'''
```

```python
# Q8. What is the binomial distribution, and how is it used in probability?

'''
The binomial distribution describes the number of successes in a fixed number of independent Bernoulli trials. 
It is used to model situations where there are a fixed number of trials, each with the same probability of success.

Example: The number of heads in 10 coin flips follows a binomial distribution.

Explanation: The binomial distribution calculates the probability of getting a certain number of successes (e.g., 
heads) in a fixed number of trials (e.g., 10 coin flips).
'''
```

```python
# Q9. What is the Poisson distribution, and where is it applied?

'''
The Poisson distribution describes the number of events occurring within a fixed interval of time or space. It is
used to model rare events.

Example: The number of emails received in an hour follows a Poisson distribution.
Explanation: The Poisson distribution calculates the probability of a given number of events (e.g., emails) occurring
in a fixed interval (e.g., one hour).
'''
```

```python
# Q10. What is a continuous uniform distribution?

'''
A continuous uniform distribution is a distribution where all outcomes in a continuous range are equally likely.
The probability density function is constant within the range.

Example: The time taken for a bus to arrive within a 10-minute window follows a continuous uniform distribution.
Explanation: Any time within the 10-minute window is equally likely, and the probability density is constant.
'''
```

```python
# Q11. What are the characteristics of a normal distribution?

'''
A normal distribution is a continuous distribution characterized by its bell-shaped curve, symmetric about the
mean. It is defined by its mean (μ) and standard deviation (σ).

Example: The heights of adult men in a population follow a normal distribution.
Explanation: The normal distribution has properties such as 68% of the data falling within one standard deviation
of the mean, 95% within two standard deviations, and 99.7% within three standard deviations.
'''
```

```python
# Q12. What is the standard normal distribution, and why is it important?

'''
The standard normal distribution is a normal distribution with a mean of 0 and a standard deviation of 1. It 
is important because it allows for the standardization of normal distributions, making it easier to calculate
probabilities and compare different distributions.

Example: Converting a normal distribution to a standard normal distribution using z-scores.
Explanation: Standardizing data using z-scores allows for comparison across different normal distributions by 
converting them to a common scale.
'''
```

```python
# Q13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

'''
The Central Limit Theorem (CLT) states that the sampling distribution of the sample mean approaches a normal 
distribution as the sample size increases, regardless of the population distribution. It is critical because it
allows for the use of normal distribution techniques in hypothesis testing and confidence intervals.

Example: The average height of a large sample of students will be approximately normally distributed, even if 
the population distribution is not normal.

Explanation: The CLT enables statisticians to make inferences about population parameters using sample data,
even when the population distribution is unknown.
'''
```

```python
# Q14. How does the Central Limit Theorem relate to the normal distribution?

'''
The Central Limit Theorem relates to the normal distribution by stating that the distribution of the sample 
mean will be approximately normal, regardless of the population distribution, as the sample size increases.

Example: The distribution of the average test scores of a large sample of students will be approximately normal.

Explanation: The CLT ensures that the sample mean follows a normal distribution, allowing for the application of 
normal distribution techniques in statistical analysis.
'''
```

```python
# Q15. What is the application of Z statistics in hypothesis testing?

'''
Z statistics are used in hypothesis testing to determine whether to reject the null hypothesis. It involves 
calculating the z-score and comparing it to a critical value from the standard normal distribution.

Example: Testing whether the mean of a sample is significantly different from a known population mean.

Explanation: The z-score measures how many standard deviations an observation is from the mean, and it is used to
assess the significance of the difference between sample and population means.
'''
```

```python
# Q16. How do you calculate a Z-score, and what does it represent?

'''
A Z-score is calculated by subtracting the mean from the observed value and dividing by the standard deviation.
It represents the number of standard deviations an observation is from the mean. Formula: Z = (X - μ) / σ

Example: A test score of 85 with a mean of 75 and a standard deviation of 10 has a z-score of (85 - 75) / 10 = 1.
Explanation: The z-score indicates how far the observed value is from the mean in terms of standard deviations.
'''
```

```python
# Q17. What are point estimates and interval estimates in statistics?

'''
Point estimates provide a single value as an estimate of a population parameter, while interval estimates provide
a range of values within which the parameter is expected to lie.

Example: The sample mean is a point estimate of the population mean, while a confidence interval is an interval
estimate.
Explanation: Point estimates give a specific value, while interval estimates provide a range that accounts for
uncertainty and variability in the data.
'''
```

```python
# Q18. What is the significance of confidence intervals in statistical analysis?

'''
Confidence intervals provide a range of values within which the true population parameter is expected to lie 
with a certain level of confidence. They give an indication of the precision and reliability of the estimate.

Example: A 95% confidence interval for the mean height of students might be 170 cm to 180 cm.
Explanation: Confidence intervals account for sampling variability and provide a range that is likely to contain
the true population parameter.
'''
```

```python
# Q19. What is the relationship between a Z-score and a confidence interval?

'''
The Z-score is used to calculate the margin of error for a confidence interval. The confidence interval is 
then constructed by adding and subtracting the margin of error from the point estimate.

Example: For a 95% confidence interval, the z-score is 1.96. The margin of error is 1.96 times the standard 
error.
Explanation: The z-score determines the width of the confidence interval, which reflects the level of confidence
in the estimate.
'''
```

```python
# Q20. How are Z-scores used to compare different distributions?

'''
Z-scores standardize different distributions, allowing for comparison by converting values to a common scale.
This makes it possible to compare observations from different distributions.

Example: Comparing test scores from different exams with different means and standard deviations.
Explanation: Z-scores enable comparison by converting values to a standard normal distribution, facilitating 
meaningful comparisons across different datasets.
'''
```

```python
# Q21. What are the assumptions for applying the Central Limit Theorem?

'''
The assumptions for applying the Central Limit Theorem are:
The samples must be independent.
The sample size should be sufficiently large (usually n > 30).
The population distribution should not be extremely skewed.
'''
```

```python
# Q22. What is the concept of expected value in a probability distribution?

'''
The expected value is the long-term average or mean value of a random variable. It is calculated by summing the
products of each possible value and its probability. Formula: E(X) = Σ [x * P(x)]

Example: The expected value of rolling a fair six-sided die is (1+2+3+4+5+6)/6 = 3.5.
Explanation: The expected value represents the average outcome if the random experiment is repeated many times.
'''
```

```python
# Q23. How does a probability distribution relate to the expected outcome of a random variable?

'''
A probability distribution describes the likelihood of different outcomes for a random variable. The expected 
outcome is the mean value of the random variable, calculated using the probability distribution.

Example: The expected number of heads in 10 coin flips is 10 * 0.5 = 5.
Explanation: The probability distribution provides the probabilities of different outcomes, and the expected 
outcome is the weighted average of these outcomes.
'''
```

```python
# Q2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

import numpy as np
import matplotlib.pyplot as plt

def discrete_uniform_distribution(low, high, size):
  return np.random.randint(low, high + 1, size)

def plot_pmf(data):
    values, counts = np.unique(data, return_counts=True)
    probabilities = counts / len(data)

    plt.figure(figsize=(8, 6))
    plt.bar(values, probabilities)
    plt.xlabel("Value")
    plt.ylabel("Probability")
    plt.title("Probability Mass Function (PMF)")
    plt.show()

low = 1
high = 6
size = 1000
uniform_data = discrete_uniform_distribution(low, high, size)

plot_pmf(uniform_data)
```

```python
# Q3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

def bernoulli_pdf(x, p):
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1")

    return p**x * (1 - p)**(1 - x)

p = 0.6
print(f"P(X=0) = {bernoulli_pdf(0, p)}")
print(f"P(X=1) = {bernoulli_pdf(1, p)}")
```

```python
# Q14. Write a Python program to calculate the Z-score for a given data point and compare it to a 
# standard normal distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data, value):
    mean = np.mean(data)
    std_dev = np.std(data)
    z_score = (value - mean) / std_dev
    return z_score

np.random.seed(0)
data = np.random.normal(0, 1, 1000)

value = 1.5
z_score = calculate_z_score(data, value)
print(f"Z-score for the value {value}: {z_score}")

x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Standard Normal Distribution', color='blue')
plt.axvline(z_score, color='red', linestyle='--', label=f'Z-score = {z_score:.2f}')

plt.title('Standard Normal Distribution and Z-score')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')

plt.grid(True)
plt.legend()

plt.show()
```

```python
# Q15. Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
import scipy.stats as stats

np.random.seed(0)
population_mean = 50
population_std = 10
sample_size = 30
sample_data = np.random.normal(population_mean, population_std, sample_size)

sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
population_mean = 50

standard_error = sample_std / np.sqrt(sample_size)

z_statistic = (sample_mean - population_mean) / standard_error
print(f"Z-statistic: {z_statistic}")

p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))
print(f"P-value: {p_value}")

alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis (H0).")
else:
    print("Fail to reject the null hypothesis (H0).")
```

```python
# Q16. Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

np.random.seed(0)
population_mean = 50
population_std = 10
sample_size = 30
sample_data = np.random.normal(population_mean, population_std, sample_size)

sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
standard_error = sample_std / np.sqrt(sample_size)

confidence_level = 0.95
z_score = stats.norm.ppf((1 + confidence_level) / 2)

margin_of_error = z_score * standard_error
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print(f"Sample Mean: {sample_mean}")
print(f"Standard Error: {standard_error}")
print(f"Z-score: {z_score}")
print(f"Margin of Error: {margin_of_error}")
print(f"Confidence Interval: {confidence_interval}")

print(f"With {confidence_level*100}% confidence, the population mean lies between {confidence_interval[0]:.2f} and {confidence_interval[1]:.2f}.")

plt.figure(figsize=(10, 6))
plt.hist(sample_data, bins=10, edgecolor='k', alpha=0.7, density=True)
plt.axvline(sample_mean, color='red', linestyle='--', label=f'Sample Mean: {sample_mean:.2f}')
plt.axvline(confidence_interval[0], color='green', linestyle='--', label=f'Lower CI: {confidence_interval[0]:.2f}')
plt.axvline(confidence_interval[1], color='green', linestyle='--', label=f'Upper CI: {confidence_interval[1]:.2f}')
plt.title('Histogram of Sample Data with Confidence Interval')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Q21. Write a Python program to compare the standard deviations of two datasets and visualize the difference.

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
data1 = np.random.normal(50, 10, 1000)  # Dataset 1: mean=50, std=10, size=1000
data2 = np.random.normal(50, 20, 1000)  # Dataset 2: mean=50, std=20, size=1000

std_dev1 = np.std(data1, ddof=1)
std_dev2 = np.std(data2, ddof=1)

print(f"Standard Deviation of Dataset 1: {std_dev1}")
print(f"Standard Deviation of Dataset 2: {std_dev2}")

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(data1, bins=30, edgecolor='k', alpha=0.7, density=True)
plt.title(f'Dataset 1\nMean: {np.mean(data1):.2f}, Std Dev: {std_dev1:.2f}')
plt.xlabel('Value')
plt.ylabel('Density')

plt.subplot(1, 2, 2)
plt.hist(data2, bins=30, edgecolor='k', alpha=0.7, density=True)
plt.title(f'Dataset 2\nMean: {np.mean(data2):.2f}, Std Dev: {std_dev2:.2f}')
plt.xlabel('Value')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
     
```

```python
# Q22. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.

import numpy as np

np.random.seed(0)
mean = 50
std_dev = 10
sample_size = 1000
data = np.random.normal(mean, std_dev, sample_size)

data_range = np.ptp(data)

q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

print(f"Range: {data_range}")
print(f"Interquartile Range (IQR): {iqr}")
```

```python
# Q23. Implement Z-score normalization on a dataset and visualize its transformation.

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
mean = 50
std_dev = 10
sample_size = 1000
data = np.random.normal(mean, std_dev, sample_size)

mean_data = np.mean(data)
std_dev_data = np.std(data)
z_score_normalized_data = (data - mean_data) / std_dev_data

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(data, bins=30, edgecolor='k', alpha=0.7, density=True)
plt.title('Original Dataset')
plt.xlabel('Value')
plt.ylabel('Density')

plt.subplot(1, 2, 2)
plt.hist(z_score_normalized_data, bins=30, edgecolor='k', alpha=0.7, density=True)
plt.title('Z-score Normalized Dataset')
plt.xlabel('Z-score')
plt.ylabel('Density')


plt.tight_layout()
plt.show()
```

```python
# Q24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a
# normal distribution.

import numpy as np
from scipy.stats import skew, kurtosis

np.random.seed(0)
mean = 50
std_dev = 10
sample_size = 1000
data = np.random.normal(mean, std_dev, sample_size)

def calculate_skewness_kurtosis(data):
    skewness = skew(data)
    kurt = kurtosis(data)
    return skewness, kurt

skewness, kurt = calculate_skewness_kurtosis(data)
print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurt}")
```
