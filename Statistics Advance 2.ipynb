```python
# Q1. What is hypothesis testing in statistics?

'''
Hypothesis testing is a statistical method used to make decisions about the properties of a population based on
a sample. It involves formulating a null hypothesis (H0) and an alternative hypothesis (H1), then using sample 
data to determine whether to reject H0.

Example: Testing whether a new drug is more effective than the current standard treatment.
'''
```

```python
# What is the null hypothesis, and how does it differ from the alternative hypothesis?

'''
The null hypothesis (H0) is a statement that there is no effect or no difference, and it serves as the default 
assumption. The alternative hypothesis (H1) is a statement that there is an effect or a difference. Hypothesis 
testing aims to determine whether there is enough evidence to reject H0 in favor of H1.
    
Example: H0: The new drug has no effect. H1: The new drug has an effect.
'''
```

```python
# Q3. What is the significance level in hypothesis testing, and why is it important?

'''
The significance level (alpha) is the probability of rejecting the null hypothesis when it is true. It is a 
threshold set by the researcher, commonly 0.05 or 0.01, to determine the strength of the evidence against H0. 
It is important because it controls the likelihood of making a Type I error.

Example: Setting alpha to 0.05 means there is a 5% chance of rejecting H0 when it is true.
'''
```

```python
# Q4. What does a P-value represent in hypothesis testing?

'''
The P-value represents the probability of obtaining a test statistic at least as extreme as the one observed, 
assuming the null hypothesis is true. It quantifies the strength of the evidence against H0.

Example: A P-value of 0.03 indicates a 3% chance of observing the data if H0 is true.
'''
```

```python
# Q5. How do you interpret the P-value in hypothesis testing?

'''
If the P-value is less than or equal to the significance level (alpha), we reject the null hypothesis. If the
P-value is greater than alpha, we fail to reject the null hypothesis. A smaller P-value indicates stronger evidence
against H0.

Example: With alpha = 0.05 , a P-value of 0.03 leads to rejecting H0, while a P-value of 0.07 leads to failing to 
reject H0.
'''
```

```python
# Q6. What are Type 1 and Type 2 errors in hypothesis testing?

'''
A Type 1 error occurs when the null hypothesis is rejected when it is true (false positive). A Type 2 error occurs 
when the null hypothesis is not rejected when it is false (false negative). The significance level (alpha) controls 
the probability of a Type 1 error, while the power of the test affects the probability of a Type 2 error.

Example: Type 1 error: Concluding a new drug is effective when it is not. Type 2 error: Concluding a new drug is not 
effective when it is.
'''
```

```python
# Q7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

'''
A one-tailed test evaluates the effect in one direction (either greater than or less than), while a two-tailed test
evaluates the effect in both directions (both greater than and less than). The choice depends on the research 
question.

Example: One-tailed test: Testing if a new drug is more effective than the current treatment. Two-tailed test:
Testing if a new drug is either more or less effective than the current treatment.
'''
```

```python
# Q8. What is the Z-test, and when is it used in hypothesis testing?

'''
The Z-test is a statistical test used to determine whether there is a significant difference between the sample
mean and the population mean when the population variance is known. It is used when the sample size is large (n > 30).

Example: Testing if the average height of a sample of students differs from the known average height of the population.
'''
```

```python
# Q9. How do you calculate the Z-score, and what does it represent in hypothesis testing?

'''
The Z-score is calculated by subtracting the population mean from the sample mean and dividing by the standard error. 
It represents the number of standard deviations the sample mean is from the population mean. Formula: Z = (X̄ - μ) / 
(σ / √n)

Example: A Z-score of 2 indicates the sample mean is 2 standard deviations above the population mean.
'''
```

```python
# Q10. What is the T-distribution, and when should it be used instead of the normal distribution?

'''
The T-distribution is a probability distribution used when the sample size is small (n ≤ 30) and the population 
variance is unknown. It is similar to the normal distribution but has heavier tails. It should be used instead of 
the normal distribution when these conditions are met.

Example: Estimating the mean of a small sample of test scores when the population variance is unknown.
'''
```

```python
# Q11. What is the difference between a Z-test and a T-test?

'''
The Z-test is used when the population variance is known and the sample size is large, while the T-test is used 
when the population variance is unknown and the sample size is small. The T-test uses the T-distribution, which
accounts for the additional uncertainty in estimating the population variance.

Example: Z-test: Testing the mean of a large sample with known population variance. T-test: Testing the mean of a
small sample with unknown population variance.
'''
```

```python
# Q12. What is the T-test, and how is it used in hypothesis testing?

'''
The T-test is a statistical test used to determine whether there is a significant difference between the means 
of two groups or between a sample mean and a population mean when the population variance is unknown. It is used
in various forms, such as the one-sample T-test, independent T-test, and paired T-test.
    
Example: Comparing the average test scores of two different classes to see if there is a significant difference.
'''
```

```python
# Q13. What is the relationship between Z-test and T-test in hypothesis testing?

'''
Both the Z-test and T-test are used to test hypotheses about population means. The Z-test is used when the 
population variance is known and the sample size is large, while the T-test is used when the population variance
is unknown and the sample size is small. As the sample size increases, the T-distribution approaches the normal 
distribution, making the T-test similar to the Z-test.

Example: For large samples, the results of the Z-test and T-test will be similar.
'''
```

```python
# Q14. What is a confidence interval, and how is it used to interpret statistical results?

'''
A confidence interval is a range of values within which the true population parameter is expected to lie with
a certain level of confidence (e.g., 95%). It provides an estimate of the precision and reliability of the 
sample statistic. Confidence intervals are used to interpret statistical results by indicating the range of 
plausible values for the population parameter.

Example: A 95% confidence interval for the mean height of students might be 170 cm to 180 cm, indicating that 
we are 95% confident that the true mean lies within this range.
'''
```

```python
# Q15. What is the margin of error, and how does it affect the confidence interval?

'''
The margin of error is the amount added and subtracted from the sample statistic to create the confidence 
interval. It reflects the precision of the estimate and is influenced by the sample size and variability. 
A larger margin of error results in a wider confidence interval, indicating less precision.

Example: If the sample mean is 175 cm and the margin of error is 5 cm, the confidence interval is 170 cm to 
180 cm.
'''
```

```python
# Q16. How is Bayes' Theorem used in statistics, and what is its significance?

'''
Bayes' Theorem is used to update the probability of a hypothesis based on new evidence. It combines prior
probability with the likelihood of the observed data to calculate the posterior probability. It is significant 
because it provides a framework for incorporating new information and making probabilistic inferences.

Example: Updating the probability of a disease given a positive test result, considering the prior probability 
of the disease and the accuracy of the test.
'''
```

```python
# Q17. What is the Chi-square distribution, and when is it used?

'''
The Chi-square distribution is a probability distribution used in hypothesis testing for categorical data. It
is used to test the independence of two variables (Chi-square test of independence) or the goodness of fit of 
an observed distribution to an expected distribution (Chi-square goodness of fit test).

Example: Testing whether the distribution of colors in a bag of candies matches the expected distribution.
'''
```

```python
# Q18. What is the Chi-square goodness of fit test, and how is it applied?

'''
The Chi-square goodness of fit test is used to determine whether an observed frequency distribution matches 
an expected distribution. It compares the observed frequencies with the expected frequencies and calculates 
the Chi-square statistic to assess the fit.

Example: Testing whether the observed frequencies of different types of fruits in a basket match the expected
frequencies based on market data.
'''
```

```python
# Q19. What is the F-distribution, and when is it used in hypothesis testing?

'''
The F-distribution is a probability distribution used in hypothesis testing for comparing variances. It is
used in the analysis of variance (ANOVA) to test whether the variances of two or more groups are equal. The
F-distribution is right-skewed and depends on the degrees of freedom of the numerator and denominator.

Example: Testing whether the variances of test scores in three different classes are equal using ANOVA.
'''
```

```python
# Q20. What is an ANOVA test, and what are its assumptions?

'''
ANOVA (Analysis of Variance) is a statistical test used to compare the means of three or more groups to 
determine if there are significant differences among them. The assumptions of ANOVA include independence 
of observations, normality of the data, and homogeneity of variances.

Example: Comparing the average test scores of students in different teaching methods to see if there are 
significant differences.
'''
```

```python
# Q21. What are the different types of ANOVA tests?

'''
There are several types of ANOVA tests, including one-way ANOVA (comparing means of three or more groups based
on one factor), two-way ANOVA (comparing means based on two factors), and repeated measures ANOVA (comparing 
means of the same subjects under different conditions).

Example: One-way ANOVA: Comparing test scores of students in different classes. Two-way ANOVA: Comparing test
scores based on teaching method and gender.
'''
```

```python
Q22. What is the F-test, and how does it relate to hypothesis testing?

'''
The F-test is a statistical test used to compare the variances of two or more groups. It is used in ANOVA to 
determine whether the group means are significantly different. The F-test calculates the ratio of the variances
and compares it to the F-distribution to assess significance.

Example: Using the F-test in ANOVA to compare the variances of test scores in different classes and determine 
if there are significant differences among the means.
'''
```

```python
# Q2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

import numpy as np
import matplotlib.pyplot as plt

def discrete_uniform_distribution(low, high, size):
  return np.random.randint(low, high + 1, size)

def plot_pmf(data):
    values, counts = np.unique(data, return_counts=True)
    probabilities = counts / len(data)

    plt.figure(figsize=(8, 6))
    plt.bar(values, probabilities)
    plt.xlabel("Value")
    plt.ylabel("Probability")
    plt.title("Probability Mass Function (PMF)")
    plt.show()

low = 1
high = 6
size = 1000
uniform_data = discrete_uniform_distribution(low, high, size)

plot_pmf(uniform_data)
```

```python
# Q3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

def bernoulli_pdf(x, p):
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1")

    return p**x * (1 - p)**(1 - x)

p = 0.6
print(f"P(X=0) = {bernoulli_pdf(0, p)}")
print(f"P(X=1) = {bernoulli_pdf(1, p)}")
```

```python
# Q15. Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
import scipy.stats as stats

np.random.seed(0)
population_mean = 50
population_std = 10
sample_size = 30
sample_data = np.random.normal(population_mean, population_std, sample_size)

sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
population_mean = 50

standard_error = sample_std / np.sqrt(sample_size)

z_statistic = (sample_mean - population_mean) / standard_error
print(f"Z-statistic: {z_statistic}")

p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))
print(f"P-value: {p_value}")

alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis (H0).")
else:
    print("Fail to reject the null hypothesis (H0).")
```

```python
# Q16. Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

np.random.seed(0)
population_mean = 50
population_std = 10
sample_size = 30
sample_data = np.random.normal(population_mean, population_std, sample_size)

sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
standard_error = sample_std / np.sqrt(sample_size)

confidence_level = 0.95
z_score = stats.norm.ppf((1 + confidence_level) / 2)

margin_of_error = z_score * standard_error
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print(f"Sample Mean: {sample_mean}")
print(f"Standard Error: {standard_error}")
print(f"Z-score: {z_score}")
print(f"Margin of Error: {margin_of_error}")
print(f"Confidence Interval: {confidence_interval}")

print(f"With {confidence_level*100}% confidence, the population mean lies between {confidence_interval[0]:.2f} and {confidence_interval[1]:.2f}.")

plt.figure(figsize=(10, 6))
plt.hist(sample_data, bins=10, edgecolor='k', alpha=0.7, density=True)
plt.axvline(sample_mean, color='red', linestyle='--', label=f'Sample Mean: {sample_mean:.2f}')
plt.axvline(confidence_interval[0], color='green', linestyle='--', label=f'Lower CI: {confidence_interval[0]:.2f}')
plt.axvline(confidence_interval[1], color='green', linestyle='--', label=f'Upper CI: {confidence_interval[1]:.2f}')
plt.title('Histogram of Sample Data with Confidence Interval')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Q1. Write a Python program to perform a Z-test for comparing a sample mean to a known population
# mean and interpret the results.

import numpy as np
import scipy.stats as stats

np.random.seed(0)
population_mean = 50
population_std = 10
sample_size = 30
sample_data = np.random.normal(population_mean, population_std, sample_size)

sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
population_mean = 50

standard_error = sample_std / np.sqrt(sample_size)

z_statistic = (sample_mean - population_mean) / standard_error
print(f"Z-statistic: {z_statistic}")

p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))
print(f"P-value: {p_value}")

alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis (H0).")
    print("There is sufficient evidence to conclude that the sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis (H0).")
    print("There is not sufficient evidence to conclude that the sample mean is significantly different from the population mean.")
```

```python
# Q3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean.

import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample_data, population_mean, population_std):

    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    p_value = 2 * (1 - norm.cdf(np.abs(z_score)))

    alpha = 0.05
    result = "Fail to reject the null hypothesis"
    if p_value < alpha:
        result = "Reject the null hypothesis"

    return {"z_score": z_score, "p_value": p_value, "result": result}

sample_data = [65, 70, 72, 68, 75, 78, 60, 71, 66, 73]
population_mean = 70
population_std = 5

test_result = one_sample_z_test(sample_data, population_mean, population_std)
test_result
```

```python
# Q4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def perform_two_tailed_z_test(sample_mean, population_mean, population_std, sample_size, alpha):

    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    critical_value = norm.ppf(1 - alpha / 2)
    print(f"Z-score: {z_score}")
    print(f"Critical Value: ±{critical_value}")

    x = np.linspace(-4, 4, 500)
    y = norm.pdf(x, 0, 1)
    plt.plot(x, y)

    x_reject_left = np.linspace(-4, -critical_value, 100)
    y_reject_left = norm.pdf(x_reject_left, 0, 1)
    plt.fill_between(x_reject_left, y_reject_left, color='red', alpha=0.3)

    x_reject_right = np.linspace(critical_value, 4, 100)
    y_reject_right = norm.pdf(x_reject_right, 0, 1)
    plt.fill_between(x_reject_right, y_reject_right, color='red', alpha=0.3)

    plt.plot(z_score, 0, 'bo', markersize=8)

    plt.title('Two-tailed Z-test')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.grid(True)
    plt.show()

    if abs(z_score) > critical_value:
        print("Reject the null hypothesis.")
    else:
        print("Fail to reject the null hypothesis.")


sample_mean = 72
population_mean = 70
population_std = 10
sample_size = 30
alpha = 0.05
perform_two_tailed_z_test(sample_mean, population_mean, population_std, sample_size, alpha)
```

```python
# Q5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

def visualize_errors(true_positive_rate, false_positive_rate, title="Type I and Type II Errors"):
    type_1_error = false_positive_rate
    type_2_error = 1 - true_positive_rate

    plt.figure(figsize=(8, 6))

    plt.bar(['Type I Error', 'Type II Error'], [type_1_error, type_2_error], color=['red', 'blue'])

    plt.title(title)
    plt.ylabel('Error Rate')
    plt.ylim(0, 1)

    for i, v in enumerate([type_1_error, type_2_error]):
        plt.text(i, v + 0.02, f"{v:.2f}", ha='center')

    plt.show()

true_positive_rate = 0.85  #true positive rate 85%
false_positive_rate = 0.10 #false positive rate 10%

visualize_errors(true_positive_rate, false_positive_rate)
```

```python
# Q6. Write a Python program to perform an independent T-test and interpret the results.

from scipy.stats import ttest_ind

def perform_independent_t_test(group1, group2):
    t_statistic, p_value = ttest_ind(group1, group2)

    interpretation = ""
    alpha = 0.05

    if p_value < alpha:
        interpretation = "There is a statistically significant difference between the two groups."
    else:
        interpretation = "There is no statistically significant difference between the two groups."

    return {
        "t-statistic": t_statistic,
        "p-value": p_value,
        "interpretation": interpretation
    }

group1 = [10, 12, 15, 14, 16]
group2 = [8, 9, 11, 10, 13]

results = perform_independent_t_test(group1, group2)
results
```

```python
# Q7. Perform a paired sample T-test using Python and visualize the comparison results.

import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt

def paired_t_test_and_visualization(data1, data2, alpha=0.05):
    t_statistic, p_value = stats.ttest_rel(data1, data2)

    print(f"T-statistic: {t_statistic}")
    print(f"P-value: {p_value}")

    if p_value < alpha:
      print("Reject the null hypothesis. There is a significant difference between the two groups.")
    else:
      print("Fail to reject the null hypothesis. There is no significant difference between the two groups.")

    plt.figure(figsize=(8, 6))

    df = pd.DataFrame({'Group1': data1, 'Group2': data2})

    plt.subplot(1,2,1)
    df.boxplot()
    plt.title('Box Plot Comparison')
    plt.ylabel('Values')

    plt.subplot(1,2,2)
    plt.scatter(data1, data2)
    plt.xlabel('Group 1')
    plt.ylabel('Group 2')
    plt.title('Scatter Plot of Paired Data')
    plt.plot([min(data1), max(data1)], [min(data1), max(data1)], color='red', linestyle='--') # Add diagonal line

    plt.tight_layout()
    plt.show()

data_group1 = [5, 10, 12, 8, 7]
data_group2 = [7, 11, 14, 9, 8]

paired_t_test_and_visualization(data_group1, data_group2)
```

```python
# Q9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import math

def confidence_interval(data, confidence=0.95):
    if not isinstance(data, (list, np.ndarray)):
        print("Error: Input data must be a list or numpy array.")
        return None

    if len(data) < 2 :
        print("Error: Input data must have at least 2 data points.")
        return None

    n = len(data)
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)

    alpha = 1 - confidence
    z_critical = norm.ppf(1 - alpha / 2)

    margin_of_error = z_critical * (sample_std / math.sqrt(n))

    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

data = [65, 70, 72, 68, 75, 78, 62, 74, 71, 69]
confidence_level = 0.95

interval = confidence_interval(data, confidence_level)

if interval:
    lower, upper = interval
    print(f"{confidence_level * 100:.0f}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

```python
# Q10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

import numpy as np
from scipy.stats import t

def margin_of_error(sample_data, confidence_level):
    n = len(sample_data)
    if n < 2:
        raise ValueError("Sample size must be at least 2.")

    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)

    alpha = 1 - confidence_level
    degrees_of_freedom = n - 1

    t_critical = t.ppf(1 - alpha / 2, degrees_of_freedom)
    margin_error = t_critical * (sample_std / np.sqrt(n))

    return margin_error

sample_data = [70, 75, 68, 72, 78, 70, 73]
confidence_level = 0.95

margin_error_value = margin_of_error(sample_data, confidence_level)
print(f"The margin of error at {confidence_level*100:.0f}% confidence is: {margin_error_value:.2f}")
```

```python
# Q11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

import numpy as np
from scipy.stats import norm

def bayesian_inference(prior_probability, likelihood, evidence):
    if evidence == 0:
        print("Error: Evidence probability cannot be zero.")
        return None
    posterior_probability = (prior_probability * likelihood) / evidence
    return posterior_probability

prior_probability = 0.01

likelihood = 0.95

true_positive_rate = 0.95
false_positive_rate = 0.05
evidence = (prior_probability * true_positive_rate) + ((1 - prior_probability) * false_positive_rate)

posterior_probability = bayesian_inference(prior_probability, likelihood, evidence)

if posterior_probability is not None:
    print(f"Posterior probability of having the disease given a positive test: {posterior_probability:.4f}")
```

```python
# Q12. Perform a Chi-square test for independence between two categorical variables in Python.

import pandas as pd
from scipy.stats import chi2_contingency

def perform_chi_square_test(data, variable1, variable2):
    try:
        contingency_table = pd.crosstab(data[variable1], data[variable2])
        chi2, p, dof, expected = chi2_contingency(contingency_table)

        return {
            "chi2_statistic": chi2,
            "p_value": p,
            "degrees_of_freedom": dof,
            "expected_frequencies": expected
        }
    except KeyError:
        return "Error: One or both of the specified variables are not found in the DataFrame."
    except TypeError:
        return "Error: Invalid input data. Ensure both variables are categorical."

data = {'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
        'Smoker': ['Yes', 'No', 'Yes', 'Yes', 'No', 'No', 'Yes', 'No']}
df = pd.DataFrame(data)

results = perform_chi_square_test(df, 'Gender', 'Smoker')

if isinstance(results, dict):
    print(f"Chi-square statistic: {results['chi2_statistic']}")
    print(f"P-value: {results['p_value']}")
    print(f"Degrees of freedom: {results['degrees_of_freedom']}")
    print(f"Expected frequencies:\n{results['expected_frequencies']}")
else:
    print(results)
```

```python
# Q13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.

import numpy as np

def calculate_expected_frequencies(observed_frequencies, total_observations):
    if not isinstance(observed_frequencies, (list, np.ndarray)):
        print("Error: observed_frequencies must be a list or numpy array.")
        return None

    if not all(isinstance(freq, (int, float)) and freq >= 0 for freq in observed_frequencies):
        print("Error: observed frequencies must be non-negative numbers.")
        return None

    if not isinstance(total_observations, (int, float)) or total_observations <= 0:
      print("Error: total_observations must be a positive number.")
      return None

    observed_sum = sum(observed_frequencies)
    if observed_sum != total_observations:
      print("Warning: The sum of observed frequencies does not equal total_observations.")

    expected_frequencies = np.array(observed_frequencies) / observed_sum * total_observations
    return expected_frequencies

observed = [10, 20, 15, 5]
total = 50
expected = calculate_expected_frequencies(observed, total)

if expected is not None:
    print("Expected frequencies:", expected)
```

```python
# Q14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

from scipy.stats import chisquare

observed_data = [10, 15, 20, 25, 30]

expected_distribution = [12, 18, 20, 22, 28]

chi2_statistic, p_value = chisquare(observed_data, expected_distribution)

print(f"Chi-square statistic: {chi2_statistic}")
print(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("The observed data does not fit the expected distribution (reject the null hypothesis).")
else:
    print("The observed data fits the expected distribution (fail to reject the null hypothesis).")
```
